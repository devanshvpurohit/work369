### Model Evaluation Metrics: Accuracy, Error Rate, Precision, Recall, and F1-Score

#### 1. Accuracy (Classification)
Accuracy measures the percentage of correct predictions out of the total number of cases. It is defined as:

$$\text{Accuracy} = \frac{\text{Number of Correct Predictions}}{\text{Total Number of Predictions}}$$

#### 2. Error Rate (Classification)
Error Rate is the proportion of incorrect predictions made by the model. It is the complement of Accuracy.

$$\text{Error Rate} = 1 - \text{Accuracy} = \frac{\text{Number of Incorrect Predictions}}{\text{Total Number of Predictions}}$$

#### 3. Precision (Classification)
Precision measures the proportion of true positive predictions among all positive predictions made by the model. It is defined as:

$$\text{Precision} = \frac{\text{True Positives}}{\text{True Positives} + \text{False Positives}}$$

#### 4. Recall (Classification)
Recall measures the proportion of true positive predictions among all actual positive instances. It is defined as:

$$\text{Recall} = \frac{\text{True Positives}}{\text{True Positives} + \text{False Negatives}}$$

#### 5. F1-Score (Classification)
The F1-score is the harmonic mean of Precision and Recall. It's a useful metric when you need a balance between Precision and Recall, especially in cases of uneven class distribution.

$$\text{F1-Score} = 2 \times \frac{\text{Precision} \times \text{Recall}}{\text{Precision} + \text{Recall}}$$

#### 6. Mean Squared Error (MSE) (Regression)
MSE measures the average of the squares of the errors—that is, the average squared difference between the estimated values ($\hat{y}$) and the actual value ($y$). It is defined as:

$$\text{MSE} = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2$$
Lower values of MSE indicate better model performance.

## Deep Dive: Portable Recursive Language Model (P-RLM) Project

This project introduces the **Portable Recursive Language Model (P-RLM)**, a novel prototype designed to tackle the limitations of traditional Retrieval-Augmented Generation (RAG) systems when dealing with extremely large and complex documents (1M+ tokens). P-RLM simulates a sophisticated, recursive reasoning agent, enabling it to navigate, understand, and synthesize information from vast textual environments more effectively than conventional methods.

### Core Design Principles and Features:

1.  **Recursive Planning**: At its heart, P-RLM employs a recursive planner that breaks down complex user queries into a series of smaller, more manageable sub-tasks. This hierarchical decomposition allows the system to approach multi-step reasoning problems systematically, much like a human would.

2.  **Portable Context**: A crucial component is the `PortableContext`, which acts as a persistent memory or 'backpack' for the reasoning agent. It diligently tracks intermediate results, stores variables and findings, logs every reasoning step, and keeps a record of visited document segments. This continuous context passing is vital for maintaining coherence and building knowledge across recursive calls.

3.  **Smart Retrieval (Chunk-based Environment)**: The `PromptEnvironment` simulates a large document by dividing it into manageable chunks. P-RLM interacts with this environment using intelligent retrieval mechanisms, including built-in caching (`_cache`) to minimize redundant searches. This ensures that the system focuses its attention on the most relevant parts of the document at each step.

4.  **Asynchronous Execution**: Leveraging `asyncio.gather`, P-RLM can execute multiple sub-tasks in parallel, significantly improving efficiency without sacrificing the depth of its recursive exploration.

5.  **LLM Integration (Simulated)**: While the prototype uses a simulated LLM wrapper (`llm_call`) for demonstration, it's designed for seamless integration with real LLM APIs (e.g., OpenAI, Gemini). This wrapper injects the `PortableContext` into prompts, allowing the LLM to leverage accumulated knowledge.

### P-RLM's Superiority Over Standard RAG Systems:

P-RLM addresses several critical shortcomings of conventional RAG, particularly in complex scenarios:

*   **Multi-Hop Reasoning**: Unlike standard RAG, which often fails when an answer requires piecing together information from multiple, non-obvious sources (e.g., finding 'Fact B' requires 'Fact A' first), P-RLM's recursive nature allows it to discover 'Fact A', store it in its `PortableContext`, and then use that new information to locate 'Fact B'. This is exemplified in the 'Secret Key' problem.

*   **Global Document Synthesis**: RAG typically returns isolated text fragments that are locally similar to a query, often failing to capture overarching themes or to synthesize a complete picture from widely distributed facts. P-RLM's **Aggregation** step, a core part of its recursive loop, actively summarizes findings from multiple branches, enabling it to construct coherent, global answers.

*   **High-Precision / Low-Noise Requirements**: Standard RAG can struggle with noisy or irrelevant retrieved chunks, leading to LLM hallucinations. P-RLM's dynamic filtering through its **Planner** ensures that only highly relevant, focused segments are passed into the reasoning process, significantly reducing noise and improving precision.

*   **Handling 'Lost in the Middle' Scenarios**: Large context windows in RAG models can lead to information in the middle of a prompt being overlooked. P-RLM mitigates this by treating the document as an navigable environment rather than a single prompt. It maintains high focus on specific, relevant chunks regardless of their original position, ensuring critical information is not 'lost'.

### Sensitivity Analysis: Impact of `max_depth` on P-RLM Performance

In this section, we evaluate how the maximum recursion depth affects the model's ability to solve complex tasks, its latency, and the number of internal logic steps required.

    *   **RAG**: Often retrieves fragments that are contextually similar but misses the 'big picture' or overall trends across 1M+ tokens.
    *   **P-RLM**: Through the **Aggregation** step, it summarizes findings from multiple branches, creating a coherent global answer.

3.  **High-Precision / Low-Noise Requirements**:
    *   **RAG**: If the top-K retrieved chunks are irrelevant, the LLM is forced to process 'noise,' leading to hallucinations.
    *   **P-RLM**: Dynamically filters chunks through the **Planner**, ensuring only highly relevant segments enter the final reasoning loop.

4.  **Handling 'Lost in the Middle' Scenarios**:
    *   **RAG**: Large context windows often ignore information in the middle of a massive prompt.
    *   **P-RLM**: By treating the document as an environment to be navigated rather than a single prompt, it maintains high focus on specific chunks regardless of their position in the file.

```python
from sentence_transformers import SentenceTransformer
import faiss
import numpy as np

class RAGBaseline:
    def __init__(self, model_name='all-MiniLM-L6-v2'):
        self.model = SentenceTransformer(model_name)
        self.index = None
        self.chunks = []

    def fit(self, chunks_dict):
        """Embed all chunks and load into FAISS index."""
        self.chunks = list(chunks_dict.values())
        embeddings = self.model.encode(self.chunks)
        dimension = embeddings.shape[1]
        self.index = faiss.IndexFlatL2(dimension)
        self.index.add(np.array(embeddings).astype('float32'))
        print(f"RAG Index built with {len(self.chunks)} chunks.")

    def query(self, text, k=3):
        """Retrieve top-k relevant chunks."""
        query_embedding = self.model.encode([text])
        distances, indices = self.index.search(np.array(query_embedding).astype('float32'), k)
        return [self.chunks[i] for i in indices[0]]

async def run_rag_comparison():
    # 1. Reuse the large environment
    env = PromptEnvironment(num_paragraphs=1000, chunk_size=500)

    # 2. Setup RAG
    print("\n--- Initializing RAG Baseline ---")
    rag = RAGBaseline()
    rag.fit(env.chunks)

    # 3. Query RAG
    query = "Find information about the quick brown fox and hidden facts."
    print(f"\nQuerying RAG: '{query}'")

    start_rag = time.time()
    results = rag.query(query, k=3)
    end_rag = time.time()

    print(f"RAG Retrieval Time: {end_rag - start_rag:.4f}s")
    print("Top RAG Results:")
    for i, res in enumerate(results):
        print(f"Result {i+1}: {res[:150]}...")

# Run the comparison
import nest_asyncio
nest_asyncio.apply()
asyncio.run(run_rag_comparison())
```

#### A. Prompt Environment (Large Document Handler)

This class simulates a large document, allowing for chunk-based navigation, searching, and slicing. We'll generate a synthetic document with hidden facts for demonstration.

```python
class PromptEnvironment:
    def __init__(self, num_paragraphs=1000, chunk_size=500):
        self.num_paragraphs = num_paragraphs
        self.chunk_size = chunk_size
        self.text, self.chunks, self.hidden_facts = self._generate_large_text()
        self.chunk_ids = list(self.chunks.keys())
        # Added caching mechanism (Requirement 7)
        self._cache = {}
        print(f"Generated document with {len(self.text.split())} words and {len(self.chunks)} chunks.")
        print(f"Hidden facts injected: {len(self.hidden_facts)}")

    def _generate_large_text(self):
        paragraphs = []
        hidden_facts = {}
        lorem_ipsum_base = "Lorem ipsum dolor sit amet, consectetur adipiscing elit. Sed do eiusmod tempor incididunt ut labore et dolore magna aliqua."

        for i in range(self.num_paragraphs):
            para = f"Paragraph {i}: {lorem_ipsum_base} "
            if random.random() < 0.05:
                fact_id = str(uuid.uuid4())[:8]
                fact = f"[HIDDEN_FACT_{fact_id}: The quick brown fox jumps over the lazy dog at {time.strftime('%H:%M:%S', time.gmtime())}.]"
                para += fact
                hidden_facts[fact_id] = fact
            paragraphs.append(para)

        full_text = "\n\n".join(paragraphs)
        words = full_text.split()
        chunks = {}
        current_chunk_id = 0
        for i in range(0, len(words), self.chunk_size):
            chunk_text = ' '.join(words[i:i + self.chunk_size])
            chunks[f"chunk_{current_chunk_id}"] = chunk_text
            current_chunk_id += 1
        return full_text, chunks, hidden_facts

    def search(self, keyword):
        # Check cache before scanning
        keyword_lower = keyword.lower()
        if keyword_lower in self._cache:
            return self._cache[keyword_lower]

        matching_chunk_ids = []
        for chunk_id, text in self.chunks.items():
            if keyword_lower in text.lower():
                matching_chunk_ids.append(chunk_id)

        # Update cache
        self._cache[keyword_lower] = matching_chunk_ids
        return matching_chunk_ids

    def get_chunk(self, chunk_id):
        return self.chunks.get(chunk_id, None)

    def semantic_search(self, query):
        print(f"[PromptEnvironment] Performing simulated semantic search for: '{query}'")
        all_matching_chunk_ids = set()

        if "fact" in query.lower() or "hidden" in query.lower():
            all_matching_chunk_ids.update(self.search("[HIDDEN_FACT_"))

        keywords = query.lower().replace("[", "").replace("]", "").split()
        for keyword in keywords:
            if len(keyword) > 3:
                all_matching_chunk_ids.update(self.search(keyword))

        return sorted(list(all_matching_chunk_ids))

    def clear_cache(self):
        self._cache = {}
        print("PromptEnvironment cache cleared.")
```

#### B. Portable Context (Memory Backpack)

This class acts as a persistent memory for the RLM, storing intermediate results, tracking visited sections, and logging reasoning steps.

```python
class PortableContext:
    def __init__(self):
        self.variables = {}
        self.logs = deque()
        self.visited_chunks = set()
        self.recursive_calls = 0

    def update(self, key, value):
        """Stores or updates an intermediate result/variable."""
        self.variables[key] = value
        self.add_log(f"Context Updated: {key} = {value}")

    def add_log(self, message):
        """Adds a log message to track reasoning steps."""
        self.logs.append(f"[{time.time():.2f}] {message}")

    def mark_chunk_visited(self, chunk_id):
        """Marks a chunk as visited to avoid redundant processing."""
        if chunk_id not in self.visited_chunks:
            self.visited_chunks.add(chunk_id)
            self.add_log(f"Marked chunk as visited: {chunk_id}")

    def increment_recursion(self):
        """Increments the counter for recursive calls."""
        self.recursive_calls += 1
        self.add_log(f"Recursive call initiated. Total calls: {self.recursive_calls}")

    def __str__(self):
        return f"PortableContext(variables={self.variables}, visited_chunks={len(self.visited_chunks)}, logs_count={len(self.logs)}, recursive_calls={self.recursive_calls})"
```

#### D & E. Recursive Planner and Solver

The `plan` function uses the LLM to decompose a task, while `solve` handles the execution, including context updates and recursive branching.

```python
async def plan(task_description: str, context: PortableContext) -> list:
    """Uses the LLM to break a complex task into a list of simpler sub-tasks."""
    prompt = f"Given the task: '{task_description}', break it down into 2-3 specific sub-tasks that can be performed on document chunks."
    response = await llm_call(prompt, context, "break this task into smaller sub-tasks")

    # In a real scenario, you'd parse the LLM output.
    # For the prototype, we'll simulate returning a list of strings.
    sub_tasks = [
        f"Identify hidden facts in relevant chunks for: {task_description}",
        f"Summarize the context found regarding: {task_description}"
    ]
    context.add_log(f"Planner created {len(sub_tasks)} sub-tasks.")
    return sub_tasks

async def solve(task: str, env: PromptEnvironment, context: PortableContext, depth=0, max_depth=2):
    """Recursively solves a task by either direct LLM call or further decomposition."""
    context.increment_recursion()
    indent = "  " * depth
    print(f"{indent}[Solve Depth {depth}] Task: {task}")

    # 1. Retrieval: Find relevant chunks
    # In this prototype, we use the task keywords to find chunks
    relevant_chunk_ids = env.semantic_search(task)[:3] # Limit to 3 chunks for brevity

    if depth >= max_depth or not relevant_chunk_ids:
        # Leaf node: Perform direct extraction/summary
        results = []
        for chunk_id in relevant_chunk_ids:
            if chunk_id not in context.visited_chunks:
                content = env.get_chunk(chunk_id)
                res = await llm_call(content, context, "Identify hidden facts or summarize")
                context.mark_chunk_visited(chunk_id)
                results.append(res)

        final_leaf_res = " | ".join(results) if results else "No new info found."
        context.update(f"result_depth_{depth}_{str(uuid.uuid4())[:4]}", final_leaf_res)
        return final_leaf_res

    # 2. Recursive Step: Plan and Gather
    sub_tasks = await plan(task, context)

    # Use asyncio.gather for parallel recursion
    tasks = [solve(st, env, context, depth + 1, max_depth) for st in sub_tasks]
    sub_results = await asyncio.gather(*tasks)

    # 3. Aggregation
    aggregation_prompt = f"Sub-results: {sub_results}"
    final_output = await llm_call(aggregation_prompt, context, "Aggregate findings into a final answer")
    context.update("final_summary", final_output)

    return final_output
```

#### D & E. Recursive Planner and Solver

The `plan` function uses the LLM to decompose a task, while `solve` handles the execution, including context updates and recursive branching using `asyncio.gather`.

```python
async def plan(task_description: str, context: PortableContext) -> list:
    """Uses the LLM to break a complex task into a list of simpler sub-tasks."""
    prompt = f"Given the task: '{task_description}', break it down into 2 specific sub-tasks that can be performed on document chunks."
    response = await llm_call(prompt, context, "break this task into smaller sub-tasks")

    # In a real scenario, you'd parse the LLM output.
    # For the prototype, we simulate returning sub-tasks based on the query.
    sub_tasks = [
        f"Identify hidden facts in relevant chunks for the query: {task_description}",
        f"Summarize observations found regarding: {task_description}"
    ]
    context.add_log(f"Planner created {len(sub_tasks)} sub-tasks.")
    return sub_tasks

async def solve(task: str, env: PromptEnvironment, context: PortableContext, depth=0, max_depth=2):
    """Recursively solves a task by either direct LLM call or further decomposition."""
    context.increment_recursion()
    indent = "  " * depth
    print(f"{indent}[Solve Depth {depth}] Task: {task}")

    # 1. Retrieval: Find relevant chunks
    # We use query keywords to simulate retrieval
    relevant_chunk_ids = env.semantic_search(task)[:3]

    if depth >= max_depth or not relevant_chunk_ids:
        # Leaf node: Perform direct extraction/summary
        results = []
        for chunk_id in relevant_chunk_ids:
            if chunk_id not in context.visited_chunks:
                content = env.get_chunk(chunk_id)
                res = await llm_call(content, context, "Identify hidden facts")
                context.mark_chunk_visited(chunk_id)
                results.append(res)

        final_leaf_res = " | ".join(results) if results else "No new info found."
        context.update(f"result_depth_{depth}_{str(uuid.uuid4())[:4]}", final_leaf_res)
        return final_leaf_res

    # 2. Recursive Step: Plan and Gather
    sub_tasks = await plan(task, context)

    # Use asyncio.gather for parallel recursion (Requirement 4)
    tasks = [solve(st, env, context, depth + 1, max_depth) for st in sub_tasks]
    sub_results = await asyncio.gather(*tasks)

    # 3. Aggregation (Requirement 3.G)
    aggregation_prompt = f"Sub-results: {sub_results}"
    final_output = await llm_call(aggregation_prompt, context, "Aggregate findings into a final answer")
    context.update(f"summary_depth_{depth}", final_output)

    return final_output
```

```python
async def llm_call(prompt: str, context: PortableContext, task_description: str) -> str:
    """Simulates an LLM API call, injecting context variables into the prompt."""
    context.add_log(f"[LLM Call] Prompt initiated for task: {task_description}. Partial Prompt: {prompt[:100]}...")

    # Simulate API latency
    await asyncio.sleep(0.5)

    t_desc = task_description.lower()
    prompt_l = prompt.lower()

    # 1. Handle Planning
    if "break this task into smaller sub-tasks" in t_desc:
        return "1. Search for keys. 2. Use keys to find locations. 3. Synthesize."

    # 2. Handle Fact Extraction (The 'Secret Key' Logic)
    if "identify hidden facts" in t_desc or "identify" in t_desc:
        findings = []
        if "[key:" in prompt_l:
            start = prompt_l.find("[key:")
            end = prompt_l.find("]", start)
            findings.append(prompt[start:end+1])
        if "[location:" in prompt_l:
            start = prompt_l.find("[location:")
            end = prompt_l.find("]", start)
            findings.append(prompt[start:end+1])
        if "[hidden_fact_" in prompt_l:
            start = prompt_l.find("[hidden_fact_")
            end = prompt_l.find("]", start)
            findings.append(prompt[start:end+1])

        return " | ".join(findings) if findings else "No facts found."

    # 3. Handle Aggregation (Final Synthesis)
    if "aggregate findings" in t_desc:
        # Check if the final target 'Zurich' or 'North Sea' exists in the gathered findings
        combined_info = str(prompt_l) + str(context.variables)
        if "zurich" in combined_info:
            return "The vault is located in Zurich."
        if "north sea" in combined_info:
            return "The vault is located in the North Sea."
        return "Found various fragments but no specific location identified."

    return f"Simulated response for: {task_description}"
```
